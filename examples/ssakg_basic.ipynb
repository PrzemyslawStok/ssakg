### This example shows how to create a very small and simple SSAKG to illustrate the basic features.

We start to create a SSAKG with 10 neurons, each neuron corresponding to a single symbol.
We would like to insert sequences of length 3.


```python
from ssakg.ssakg import SSAKG
from ssakg.utils.sequence_generator import SequenceGenerator

# The graphs_to_drawing parameter is used for creation of colorful pictures of sequences to explain examples.
ssakg = SSAKG(number_of_symbols=10, sequence_length=3, graphs_to_drawing=True)
```

Let's insert first sequence and look what we have inside:

```python
ssakg.insert_sequence([1, 2, 3])
ssakg.show()
```

Let's insert another sequence, with the same symbol used in two places in a sequence. We see that another neuron (with number 10) is created for this symbol:

```python
ssakg.insert_sequence([3, 5, 5])
ssakg.show()
```

Let's insert another 100 sequences using SequenceGenerator object. Now we will see graph with many additional neurons for multiple symbols.
We can see that the part of the graph that contains multiple symbols is more sparse since this symbols appears less often.

```python
sequence_generator = SequenceGenerator(sequence_length=3, sequence_min=0, sequence_max=10)
sequences = sequence_generator.generate_unique_sequences(100)

ssakg.insert(sequences)

ssakg.show()
```
